**1 -** Define una clase de nombre `Shape`. Esta hereda dos clases: `Círculo` y `Rectángulo.` Compruebe mediante programación la relación de herencia entre las clases. Cree objetos de las clases Shape y Circle. Informe de qué clases son las instancias de estos objetos.

````python
# Definición de una clase base llamada Shape
class Shape:
    pass

# Definición de una clase llamada Rectangle que hereda de Shape
class Rectangle(Shape):
    pass

# Definición de una clase llamada Circle que hereda de Shape
class Circle(Shape):
    pass

# Creación de una instancia de la clase Shape
s = Shape()

# Creación de una instancia de la clase Circle
c = Circle()

# Verificación si s es una instancia de Shape
print(isinstance(s, Shape))      # True: s es una instancia directa de Shape

# Verificación si s es una instancia de Rectangle
print(isinstance(s, Rectangle))  # False: s es una instancia de Shape, no de Rectangle

# Verificación si s es una instancia de Circle
print(isinstance(s, Circle))     # False: s es una instancia de Shape, no de Circle

# Verificación si Rectangle es una subclase de Shape
print(issubclass(Rectangle, Shape))  # True: Rectangle hereda de Shape

# Verificación si Circle es una subclase de Shape
print(issubclass(Circle, Shape))     # True: Circle hereda de Shape

````

**2 -** Escribe un programa que utilice herencia simple entre las clases `Base` y `Derived`. Si hay un método en la clase `Base`, ¿cómo evitar que sea sobrescrito en la clase `Derived`?

````python
# Definición de la clase base llamada Base
class Base:
    # Método privado en la clase Base
    def __method(self):
        print('In Base.__method')
    
    # Método público que llama al método privado __method
    def func(self):
        self.__method()

# Definición de la clase Derived que hereda de Base
class Derived(Base):
    # Método privado en la clase Derived
    def __method(self):
        print('In Derived.__method')

# Creación de una instancia de la clase Base
b = Base()
# Llamada al método func de la instancia b
b.func()  # Esto imprimirá 'In Base.__method'

# Creación de una instancia de la clase Derived
d = Derived()
# Llamada al método func de la instancia d
d.func()  # Esto imprimirá 'In Base.__method' debido a la ocultación de nombres
````

**Tips**  
- Para evitar que se sobrescriba un método, añadir __.
- Cuando `func()` es llamado usando `b`, `self` contiene la dirección del objeto de la clase Base. Cuando es llamado usando `d`, `self` contiene la dirección del objeto de la clase `Derived`.
- En la clase Base `__method()` se convierte en `_Base__method()` y en la clase `Derived` se convierte en `_Derived__method()`.
- Cuando `func()` llama a `__method( )` desde la clase Base, es el método `_Base()` el que es llamado. En efecto, `_method()` no puede ser sobreescrito. Esto es cierto, incluso cuando `self` contiene la dirección del objeto de la clase `Derived`.

**Contexto Teórico**  
**El plano cartesiano.**  
Representa un espacio bidimensional (en 2 dimensiones), formado por dos rectas perpendiculares, una horizontal y otra vertical que se cortan en un punto. La recta horizontal se denomina eje de las abscisas o `eje X`, mientras que la vertical recibe el nombre de eje de las ordenadas o simplemente `eje Y`. En cuanto al punto donde se cortan, se conoce como el `punto de origen O`.

Es importante remarcar que el plano se divide en 4 cuadrantes:

**Puntos y coordenadas.**  

El objetivo de todo esto es describir la posición de puntos sobre el plano en forma de coordenadas, que se forman asociando el valor del eje de las X (horizontal) con el valor del eje Y (vertical).  

La representación de un punto es sencilla: P(X,Y) dónde X y la Y son la distancia horizontal (izquierda o derecha) y vertical (arriba o abajo) respectivamente, utilizando como referencia el punto de origen (0,0), justo en el centro del plano.

**Vectores en el plano**  

Finalmente, un vector en el plano hace referencia a un segmento orientado, generado a partir de dos puntos distintos.

A efectos prácticos no deja de ser una línea formada desde un punto inicial en dirección a otro punto final, por lo que se entiende que un vector tiene longitud y dirección/sentido.

En esta figura, podemos observar dos puntos A y B que podríamos definir de la siguiente forma:

- **A(x1, y1) => A(2, 3)**
  
- **B(x2, y2) => B(5, 5)**

Y el vector se representaría como la diferencia entre las coordendas del segundo punto respecto al primero (el segundo menos el primero):

**AB = (x2-x1, y2-y1) => (5-2, 5-3) => (3,2)**

Lo que en definitiva no deja de ser: 3 a la derecha y 2 arriba.

- Añade un método llamado cuadrante que indique a qué cuadrante pertenece el punto, teniendo en cuenta que si X == 0 e Y != 0 se sitúa sobre el eje Y, si X != 0 e Y == 0 se sitúa sobre el eje X y si X == 0 e Y == 0 está sobre el origen.
  
- Añade un método llamado vector, que tome otro punto y calcule el vector resultante entre los dos puntos.
  
- (Optativo) Añade un método llamado distancia, que tome otro punto y calcule la distancia entre los dos puntos y la muestre por pantalla. La fórmula es la siguiente:

- Crea una clase llamada Rectangulo con dos puntos (inicial y final) que formarán la diagonal del rectángulo.
  
- Añade un método constructor para crear ambos puntos fácilmente, si no se envían se crearán dos puntos en el origen por defecto.
  
- Añade al rectángulo un método llamado base que muestre la base.
  
- Añade al rectángulo un método llamado altura que muestre la altura.
  
- Añade al rectángulo un método llamado area que muestre el area.
  

````python
import math

# Definición de la clase Punto
class Punto:
    
    # Constructor que inicializa las coordenadas del punto
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    
    # Método especial para representar el punto como una cadena
    def __str__(self):
        return '({}, {})'.format(self.x, self.y)
    
    # Método para determinar el cuadrante del punto
    def cuadrante(self):
        if self.x > 0 and self.y > 0:
            print('El punto {} pertenece al cuadrante I '.format(self))
        elif self.x < 0 and self.y > 0:
            print('El punto {} pertenece al cuadrante II '.format(self))
        elif self.x < 0 and self.y < 0:
            print('El punto {} pertenece al cuadrante III '.format(self))
        elif self.x > 0 and self.y < 0:
            print('El punto {} pertenece al cuadrante IV '.format(self))
        else:
            print('{} se encuentra en el origen'.format(self))
    
    # Método para calcular el vector entre dos puntos
    def vector(self, p):
        print('El vector formado por {} y {} es: ({}, {})'.format(self, p, p.x - self.x, p.y - self.y))
    
    # Método para calcular la distancia entre dos puntos
    def distancia(self, p):
        d = math.sqrt((p.x - self.x)**2 + (p.y - self.y)**2)
        print('La distancia entre los puntos {} y {} es {}'.format(self, p, d))

# Definición de la clase Rectángulo
class Rectángulo:
    
    # Constructor que inicializa los puntos inicial y final del rectángulo
    def __init__(self, pInicial = Punto(), pFinal = Punto()):
        self.pInicial = pInicial
        self.pFinal = pFinal
    
    # Método para calcular y mostrar la base del rectángulo
    def base(self):
        self.base = abs(self.pFinal.x - self.pInicial.x)
        print('La base del rectángulo es {}'.format(self.base))
    
    # Método para calcular y mostrar la altura del rectángulo
    def altura(self):
        self.altura = abs(self.pFinal.y - self.pInicial.y)
        print('La altura del rectángulo es {}'.format(self.altura))
    
    # Método para calcular y mostrar el área del rectángulo
    def area(self):
        self.base = abs(self.pFinal.x - self.pInicial.x)
        self.altura = abs(self.pFinal.y - self.pInicial.y)
        self.area = self.base * self.altura
        print('El área del rectángulo es {}'.format(self.area))
        
# Creación de varios puntos
A = Punto(2,3)
B = Punto(5,5)
C = Punto(-3, -1)
D = Punto(0,0)
E = Punto(-9, 6)
F = Punto(5, 7)
G = Punto(9, -6)
H = Punto(7, 8)

####################################

# Determinación del cuadrante de los puntos
A.cuadrante()  # Imprimirá que A está en el cuadrante I
B.cuadrante()  # Imprimirá que B está en el cuadrante I
C.cuadrante()  # Imprimirá que C está en el cuadrante III

print('####################################')

# Cálculo y muestra de los vectores entre puntos
E.vector(F)  # Imprimirá el vector entre E y F
F.vector(E)  # Imprimirá el vector entre F y E

print('####################################')

# Cálculo y muestra de la distancia entre puntos
G.distancia(H)  # Imprimirá la distancia entre G y H
H.distancia(G)  # Imprimirá la distancia entre H y G

##################################################
print('###################################')

# Creación de un rectángulo usando los puntos A y B
R = Rectángulo(A, B)

# Cálculo y muestra de la base, altura y área del rectángulo
R.base()   # Imprimirá la base del rectángulo
R.altura() # Imprimirá la altura del rectángulo
R.area()   # Imprimirá el área del rectángulo
````

**4 -** En esta ejemplo vas a trabajar el concepto de herencia un poco más en profundidad, aprovechando para introducir un nuevo concepto muy importante que te facilitará la vida.  
Hasta ahora sabemos que una clase heredada puede fácilmente extender algunas funcionalidades, simplemente añadiendo nuevos atributos y métodos, o sobreescribiendo los ya existentes. Como en el siguiente ejemplo:

````python
# Definición de la clase base Vehiculo
class Vehiculo():

    # Constructor que inicializa los atributos color y ruedas
    def __init__(self, color, ruedas):
        self.color = color
        self.ruedas = ruedas

    # Método especial para representar el vehículo como una cadena
    def __str__(self):
        return "Color {}, {} ruedas".format(self.color, self.ruedas)

# Definición de la clase Coche que hereda de Vehiculo
class Coche(Vehiculo):

    # Constructor que inicializa los atributos color, ruedas, velocidad y cilindrada
    def __init__(self, color, ruedas, velocidad, cilindrada):
        # Inicialización de los atributos color y ruedas heredados de Vehiculo
        super().__init__(color, ruedas)
        self.velocidad = velocidad
        self.cilindrada = cilindrada

    # Método especial para representar el coche como una cadena
    def __str__(self):
        return "color {}, {} km/h, {} ruedas, {} cc".format(self.color, self.velocidad, self.ruedas, self.cilindrada)

# Creación de una instancia de la clase Coche
coche = Coche("azul", 4, 150, 1200)

# Impresión de la representación del objeto coche
print(coche)  # Esto imprimirá: color azul, 150 km/h, 4 ruedas, 1200 cc
````

El inconveniente más evidente de ir sobreescribiendo es que tenemos que volver a escribir el código de la superclase y luego el específico de la subclase. Para evitarnos escribir código innecesario, podemos utilizar un truco que consiste en llamar el método de la superclase y luego simplemente escribir el código de la clase:

````python
class Vehiculo():

    def __init__(self, color, ruedas):
        self.color = color
        self.ruedas = ruedas

    def __str__(self):
        return "Color {}, {} ruedas".format( self.color, self.ruedas )


class Coche(Vehiculo):

    def __init__(self, color, ruedas, velocidad, cilindrada):
        Vehiculo.__init__(self, color, ruedas)
        self.velocidad = velocidad
        self.cilindrada = cilindrada

    def __str__(self):
        return Vehiculo.__str__(self) + ", {} km/h, {} cc".format(self.velocidad, self.cilindrada)


c = Coche("azul", 4, 150, 1200)
print(c)
````

Determinar constantemente la superclase puede ser fastidioso, Python nos permite utilizar un acceso directo mucho más cómodo llamado super(). Hacerlo de esta forma nos permite llamar cómodamente los métodos o atributos de la superclase sin necesidad de especificar el self.

````python
class Vehiculo():

    def __init__(self, color, ruedas):
        self.color = color
        self.ruedas = ruedas

    def __str__(self):
        return "Color {}, {} ruedas".format( self.color, self.ruedas )


class Coche(Vehiculo):

    def __init__(self, color, ruedas, velocidad, cilindrada):
        super().__init__(color, ruedas) # Utilizanmos super() sin self en lugar de vehículo (nombre de la superclase)
        self.velocidad = velocidad
        self.cilindrada = cilindrada

    def __str__(self):
        return super().__str__() + ", {} km/h, {} cc".format(self.velocidad, self.cilindrada)


c = Coche("azul", 4, 150, 1200)
print(c)
````

#### OTRO EJEMPLO SIMILAR 

 La jerarquía de clases será la siguiente:

- Dispositivo
  - Teléfono
    - TeléfonoMóvil
    - TeléfonoFijo
  - Computadora
    - Laptop
    - Escritorio
  - Tablet


````python
class Dispositivo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def __str__(self):
        return "Marca: {}, Modelo: {}".format(self.marca, self.modelo)


class Teléfono(Dispositivo):
    def __init__(self, marca, modelo, tipo):
        super().__init__(marca, modelo)
        self.tipo = tipo

    def __str__(self):
        return "Teléfono - Marca: {}, Modelo: {}, Tipo: {}".format(self.marca, self.modelo, self.tipo)


class TeléfonoMóvil(Teléfono):
    def __init__(self, marca, modelo, sistema_operativo):
        super().__init__(marca, modelo, "Móvil")
        self.sistema_operativo = sistema_operativo

    def __str__(self):
        return "Teléfono Móvil - Marca: {}, Modelo: {}, Sistema Operativo: {}".format(self.marca, self.modelo, self.sistema_operativo)


class TeléfonoFijo(Teléfono):
    def __init__(self, marca, modelo):
        super().__init__(marca, modelo, "Fijo")

    def __str__(self):
        return "Teléfono Fijo - Marca: {}, Modelo: {}".format(self.marca, self.modelo)


class Computadora(Dispositivo):
    def __init__(self, marca, modelo, tipo):
        super().__init__(marca, modelo)
        self.tipo = tipo

    def __str__(self):
        return "Computadora - Marca: {}, Modelo: {}, Tipo: {}".format(self.marca, self.modelo, self.tipo)


class Laptop(Computadora):
    def __init__(self, marca, modelo, tamaño_pantalla):
        super().__init__(marca, modelo, "Laptop")
        self.tamaño_pantalla = tamaño_pantalla

    def __str__(self):
        return "Laptop - Marca: {}, Modelo: {}, Tamaño Pantalla: {} pulgadas".format(self.marca, self.modelo, self.tamaño_pantalla)


class Escritorio(Computadora):
    def __init__(self, marca, modelo):
        super().__init__(marca, modelo, "Escritorio")

    def __str__(self):
        return "Escritorio - Marca: {}, Modelo: {}".format(self.marca, self.modelo)


class Tablet(Dispositivo):
    def __init__(self, marca, modelo, tamaño_pantalla):
        super().__init__(marca, modelo)
        self.tamaño_pantalla = tamaño_pantalla

    def __str__(self):
        return "Tablet - Marca: {}, Modelo: {}, Tamaño Pantalla: {} pulgadas".format(self.marca, self.modelo, self.tamaño_pantalla)


# Función para catalogar dispositivos
def catalogar(dispositivos, tipo=None):
    contador = 0
    for dispositivo in dispositivos:
        if tipo is None or tipo.lower() in dispositivo.__class__.__name__.lower():
            print(dispositivo)
            contador += 1
    if tipo is not None:
        print("Se han encontrado {} dispositivos de tipo {}".format(contador, tipo))

# Crear instancias de cada subclase
dispositivos = [
    TeléfonoMóvil("Apple", "iPhone 12", "iOS"),
    TeléfonoFijo("Panasonic", "KX-TS880"),
    Laptop("Dell", "XPS 13", 13.3),
    Escritorio("HP", "Pavilion"),
    Tablet("Samsung", "Galaxy Tab S6", 10.5)
]

# Catalogar todos los dispositivos
catalogar(dispositivos)

print('###################################')

# Catalogar solo dispositivos de tipo "Teléfono"
catalogar(dispositivos, tipo="Teléfono")
````
