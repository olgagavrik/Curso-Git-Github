### Primero vamos a crear los dataframes:

```python
import pandas as pd

# DataFrame de ventas
ventas = pd.DataFrame({
    'Tienda_1': [1500, 2000, 2500, 1800, 1600],
    'Tienda_2': [1800, 2200, 2600, 2100, 1900],
    'Tienda_3': [1300, 1700, 2100, 1600, 1400],
    'Tienda_4': [2000, 2400, 2800, 2300, 2100]
}, index=['Producto_A', 'Producto_B', 'Producto_C', 'Producto_D', 'Producto_E'])

# DataFrame de costes
costes = pd.DataFrame({
    'Tienda_1': [1000, 1200, 1600, 1100, 1000],
    'Tienda_2': [1100, 1300, 1700, 1200, 1100],
    'Tienda_3': [900, 1100, 1400, 1000, 900],
    'Tienda_4': [1200, 1400, 1800, 1300, 1200]
}, index=['Producto_A', 'Producto_B', 'Producto_C', 'Producto_D', 'Producto_E'])

# DataFrame de inventario inicial
inventario_inicial = pd.DataFrame({
    'Tienda_1': [500, 700, 600, 550, 520],
    'Tienda_2': [600, 800, 700, 650, 620],
    'Tienda_3': [400, 600, 500, 450, 420],
    'Tienda_4': [700, 900, 800, 750, 720]
}, index=['Producto_A', 'Producto_B', 'Producto_C', 'Producto_D', 'Producto_E'])

# DataFrame de inventario final
inventario_final = pd.DataFrame({
    'Tienda_1': [300, 500, 400, 350, 320],
    'Tienda_2': [400, 600, 500, 450, 420],
    'Tienda_3': [200, 400, 300, 250, 220],
    'Tienda_4': [500, 700, 600, 550, 520]
}, index=['Producto_A', 'Producto_B', 'Producto_C', 'Producto_D', 'Producto_E'])

print("Ventas:")
print(ventas)
print("\nCostes:")
print(costes)
print("\nInventario Inicial:")
print(inventario_inicial)
print("\nInventario Final:")
print(inventario_final)
```

### Pregunta 1. Calcula las ganancias

```python
# Ganancias = Ventas - Costes
ganancias = ventas - costes

print("\nGanancias:")
print(ganancias)
```

### Pregunta 2. Calcular el coste de mercancía vendida (CMV)

```python
# CMV = Inventario Inicial + Compras - Inventario Final
# Asumimos que Compras es igual a Costes para simplificar
CMV = inventario_inicial + costes - inventario_final

print("\nCoste de Mercancía Vendida (CMV):")
print(CMV)
```

### Pregunta 3. Calcular el margen de beneficio

```python
# Margen de beneficio = (Ganancias / Ventas) * 100
margen_beneficio = (ganancias / ventas) * 100

print("\nMargen de Beneficio (%):")
print(margen_beneficio)
```

### Pregunta 4. Calcular el inventario promedio

```python
# Inventario Promedio = (Inventario Inicial + Inventario Final) / 2
inventario_promedio = (inventario_inicial + inventario_final) / 2

print("\nInventario Promedio:")
print(inventario_promedio)
```

### Pregunta 5. Calcular la rotación de inventario

```python
# Rotación de Inventario = CMV / Inventario Promedio
rotacion_inventario = CMV / inventario_promedio

print("\nRotación de Inventario:")
print(rotacion_inventario)
```

### Pregunta 6. Calcular el inventario final promedio por tienda

```python
# Inventario Final Promedio por Tienda
inventario_final_promedio = inventario_final.mean(axis=0)

print("\nInventario Final Promedio por Tienda:")
print(inventario_final_promedio)
```

### Pregunta 7. Analizar la relación entre rotación de inventario y ganancias

```python
# Correlación entre rotación de inventario y ganancias
correlacion_rotacion_ganancias = rotacion_inventario.corrwith(ganancias)

print("\nCorrelación entre Rotación de Inventario y Ganancias:")
print(correlacion_rotacion_ganancias)
```

### Pregunta 8. Identificar tiendas con margen de beneficio negativo

```python
# Tiendas con Margen de Beneficio Negativo
margen_negativo = margen_beneficio < 0

print("\nTiendas con Margen de Beneficio Negativo:")
print(margen_negativo)
```

### Pregunta 9. Evaluar la variabilidad de las ventas entre tiendas

```python
# Desviación Estándar de las Ventas entre Tiendas
desviacion_ventas = ventas.std(axis=1)

print("\nDesviación Estándar de las Ventas entre Tiendas:")
print(desviacion_ventas)
```

### Pregunta 10. Calcular la proporción de inventario vendido

```python
# Proporción de Inventario Vendido = (Inventario Inicial - Inventario Final) / Inventario Inicial
proporcion_inventario_vendido = (inventario_inicial - inventario_final) / inventario_inicial

print("\nProporción de Inventario Vendido:")
print(proporcion_inventario_vendido)
```

### Pregunta 11.Identificar productos con mayor variabilidad en ganancias

```python
# Desviación Estándar de las Ganancias entre Tiendas
variabilidad_ganancias = ganancias.std(axis=1)

# Productos con mayor variabilidad en ganancias
productos_mayor_variabilidad = variabilidad_ganancias.sort_values(ascending=False)

print("\nProductos con Mayor Variabilidad en Ganancias:")
print(productos_mayor_variabilidad)
```

### Pregunta 12. Comparar ventas entre productos y tiendas

```python
# Comparación de Ventas entre Productos y Tiendas
ventas_comparacion = ventas.describe()

print("\nComparación de Ventas entre Productos y Tiendas:")
print(ventas_comparacion)
```

### Pregunta 13. Analizar el impacto de los inventarios en las ganancias

```python
# Correlación entre Inventario Promedio y Ganancias
correlacion_inventario_ganancias = inventario_promedio.corrwith(ganancias)

print("\nCorrelación entre Inventario Promedio y Ganancias:")
print(correlacion_inventario_ganancias)
```

### Pregunta 14. Muestre visualizaciones de los resultados con Matplotlib

```python
import matplotlib.pyplot as plt

# Visualización de ganancias por tienda
ganancias.plot(kind='bar', title='Ganancias por Tienda')
plt.xlabel('Producto')
plt.ylabel('Ganancias')
plt.show()

# Visualización del CMV
CMV.plot(kind='bar', title='Coste de Mercancía Vendida (CMV)')
plt.xlabel('Producto')
plt.ylabel('CMV')
plt.show()

# Visualización del margen de beneficio
margen_beneficio.plot(kind='bar', title='Margen de Beneficio (%)')
plt.xlabel('Producto')
plt.ylabel('Margen de Beneficio (%)')
plt.show()

# Visualización del inventario promedio
inventario_promedio.plot(kind='bar', title='Inventario Promedio')
plt.xlabel('Producto')
plt.ylabel('Inventario Promedio')
plt.show()

# Visualización de la rotación de inventario
rotacion_inventario.plot(kind='bar', title='Rotación de Inventario', color='green')
plt.xlabel('Producto')
plt.ylabel('Rotación de Inventario')
plt.show()

# Visualización del inventario final promedio por tienda
inventario_final_promedio.plot(kind='bar', title='Inventario Final Promedio por Tienda', color='purple')
plt.xlabel('Tienda')
plt.ylabel('Inventario Final Promedio')
plt.show()

# Visualización de la correlación entre rotación de inventario y ganancias
correlacion_rotacion_ganancias.plot(kind='bar', title='Correlación entre Rotación de Inventario y Ganancias', color='orange')
plt.xlabel('Tienda')
plt.ylabel('Correlación')
plt.show()

# Visualización de la desviación estándar de las ventas entre tiendas
desviacion_ventas.plot(kind='bar', title='Desviación Estándar de las Ventas entre Tiendas', color='red')
plt.xlabel('Producto')
plt.ylabel('Desviación Estándar')
plt.show()

# Visualización de la proporción de inventario vendido
proporcion_inventario_vendido.plot(kind='bar', title='Proporción de Inventario Vendido')
plt.xlabel('Producto')
plt.ylabel('Proporción de Inventario Vendido')
plt.show()

# Visualización de la variabilidad en ganancias
productos_mayor_variabilidad.plot(kind='bar', title='Productos con Mayor Variabilidad en Ganancias')
plt.xlabel('Producto')
plt.ylabel('Desviación Estándar de Ganancias')
plt.show()

# Visualización de la comparación de ventas
ventas_comparacion.loc['mean'].plot(kind='bar', title='Comparación de Ventas entre Productos y Tiendas', color='blue')
plt.xlabel('Tienda')
plt.ylabel('Ventas Promedio')
plt.show()

# Visualización del impacto de los inventarios en las ganancias
correlacion_inventario_ganancias.plot(kind='bar', title='Correlación entre Inventario Promedio y Ganancias', color='cyan')
plt.xlabel('Producto')
plt.ylabel('Correlación')
plt.show()
```
