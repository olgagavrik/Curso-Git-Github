```python
estudiantes = {
    "Ana": {"Matemáticas": 95, "Ciencias": 90, "Historia": 85},
    "Luis": {"Matemáticas": 78, "Ciencias": 88, "Historia": 92},
    "Marta": {"Matemáticas": 88, "Ciencias": 94, "Historia": 90},
    "Juan": {"Matemáticas": 82, "Ciencias": 86, "Historia": 88},
    "Nina": {"Matemáticas": 72, "Ciencias": 95, "Historia": 80}
}

def calcular_totales_y_promedios(estudiantes):
    resultados = {}
    for nombre, calificaciones in estudiantes.items():
        total = sum(calificaciones.values())
        promedio = total / len(calificaciones)
        resultados[nombre] = {"Total": total, "Promedio": promedio}
    return resultados

def encontrar_mejor_estudiante(resultados):
    mejor_nombre = ""
    mayor_promedio = 0
    for nombre, datos in resultados.items():
        if datos["Promedio"] > mayor_promedio:
            mejor_nombre = nombre
            mayor_promedio = datos["Promedio"]
    return mejor_nombre, mayor_promedio

resultados = calcular_totales_y_promedios(estudiantes)

mejor_estudiante, mejor_promedio = encontrar_mejor_estudiante(resultados)

print("Resultados de los estudiantes:")
for nombre, datos in resultados.items():
    print(f"{nombre}: Total = {datos['Total']}, Promedio = {datos['Promedio']:.2f}")

print(f"El mejor estudiante es {mejor_estudiante} con un promedio de {mejor_promedio:.2f}.")
```

```python
interfaces = {
    "eth0": {"IP": "1.1.1.1", "Estado": "up"},
    "eth1": {"IP": "2.2.2.2", "Estado": "up"},
    "wlan0": {"IP": "3.3.3.3", "Estado": "down"},
    "wlan1": {"IP": "4.4.4.4", "Estado": "up"}
}

def estado_interfaz(interfaz):
    return interfaces.get(interfaz, {}).get("Estado", "Interfaz no encontrada")

def interfaces_activas():
    activas = {nombre: datos["IP"] for nombre, datos in interfaces.items() if datos["Estado"] == "up"}
    return activas

def numero_interfaces():
    return len(interfaces)

def anadir_entradas():
    interfaces["eth2"] = {"IP": "5.5.5.5", "Estado": "up"}
    interfaces["eth3"] = {"IP": "6.6.6.6", "Estado": "down"}

print("Estado de 'eth0':", estado_interfaz("eth0"))
print("Interfaces activas y sus IPs:", interfaces_activas())
print("Número total de interfaces:", numero_interfaces())
anadir_entradas()  # Añadir nuevas interfaces
print("Número total de interfaces después de añadir nuevas:", numero_interfaces())
```

```python
passwords = {
    'jugador': 'jugar',
    'pensador': 'pensar',
    'pescador': 'atun',
    'escritor': 'pluma',
    'escultor': 'galatea',
    'leñador': 'encina',
    'comprador': 'shopping',
    'profesor': 'chalk',
    'pecador': 'hell',
    'conductor': 'gas'
}
def verificar_credenciales(usuario, contraseña):
    if usuario in passwords and passwords[usuario] == contraseña:
        return "Acceso concedido."
    else:
        return "Acceso denegado. Usuario o contraseña incorrectos."

nombre_usuario = input("Introduce tu nombre de usuario: ")
contraseña_usuario = input("Introduce tu contraseña: ")

resultado = verificar_credenciales(nombre_usuario, contraseña_usuario)
print(resultado)
```

```python
diccionario = {
    'clave_1': ['val1', 'val2'],
    'clave_2': ['val3', 'val4', 'val5'],
    'clave_n': ['valn', 'valm', 'valr']
}

def contar_valores(diccionario):
    conteo = {}
    for clave, valores in diccionario.items():
        conteo[clave] = len(valores) 
    return conteo

resultado = contar_valores(diccionario)
print(resultado)
```

```python
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}

def combinar_diccionarios(dic1, dic2):
    
    resultado = {}
    
    todas_las_claves = set(dic1.keys()).union(dic2.keys())
    
    for clave in todas_las_claves:
        
        resultado[clave] = dic1.get(clave, 0) + dic2.get(clave, 0)
    
    return resultado

diccionario_combinado = combinar_diccionarios(d1, d2)

print(diccionario_combinado)
```
